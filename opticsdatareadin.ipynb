```r
#install.packages('reshape2', repos='http://cran.us.r-project.org')
library(reshape2)
library(ggplot2)

wdir <- "/Users/vgd/Desktop/ProSem/wavefront/VS212A_2015_Aberration_Data/ZER/" 
fn <- list.files(path=wdir, pattern="*_500.zer")
fnf <- list.files(path=wdir, pattern="*_500.zer", full.names = TRUE)

randf <- function(path){
    df <- read.table(path, sep='\t', skip=16, header=FALSE)
    df <- df[,3]
    return(df)
}

##get z1 and z2 coefficients
df <- read.table(fnf[1], sep='\t', skip=16, header=FALSE)
z1 <- df[,1]
z2 <- df[,2]
zpoly <- paste('Z',z1,z2,sep='_')


mats <- lapply(fnf, randf)
df <- data.frame(do.call(cbind.data.frame, mats))
colnames(df) <- gsub('.zer','',fn)
rownames(df) <- zpoly

df

write.table(df, file=paste(wdir,"zertable.tsv",sep=''),sep='\t',quote=FALSE)
```

```r
mn <- apply(df, 1, mean)
sd <- apply(df, 1, sd)

dfm <- melt(data.frame(zpoly, df), variable = 'measure', value.name='zcoeff', id ='zpoly')
dfm$zpoly <- factor(dfm$zpoly, levels = as.character(rownames(df)))

stat_sum_single <- function(fun, geom="point", ...) {
  stat_summary(fun.y=fun, colour="red", geom=geom, size = 2, ...)
}

pdf(paste(wdir,'meanbar.pdf',sep=''), width=15, height=5)

p <- ggplot(dfm, aes(zpoly, zcoeff))
p <- p + geom_boxplot(outlier.colour = NA)
p <- p + geom_point(position = position_jitter(width = 0.01), alpha = 1/2, size=2)
p <- p + theme(axis.text.x = element_text(angle = 45, hjust = 1))
p <- p + stat_sum_single(fun = "mean")
p

dev.off()
p
```

```r
##calculate the RMS
##compute the sum of the sqaure of all the coefficients in the list that you define.
##Compute the average RMS as a function of wavefront order 
##(2nd order includes terms 3-5, 3rd order includes terms 6-9, etc) and plot it as a bar graph***.
##Add standard error of the mean (SEM) to each bar. (***
##compute the RMS as a function of order for each individual first, then average the results for all 16 individuals, then plot)

##rms=sqrt(sum(c(1:65).^2)); 
```

```r
get.rms <- function(subj, ord){
    rms <- numeric()
    for(o in ord){
        vals <- subj[grep(paste('Z_',o,'_',sep=''),rownames(df))]
        rmsv <- sqrt(sum(vals^2))
        rms[o] <- rmsv     
    }
    return(rms)
}
ordersall <- gsub('Z_','',gsub('_[-]?[0-9]*$','',rownames(df)))
order <- unique(gsub('Z_','',gsub('_[-]?[0-9]*$','',rownames(df))))

rms <- apply(df, 2, get.rms, order)

calc.se <- function(row){
    #ruse <- row[,grep('order',colnames(row),invert=TRUE)]
    mdiff <- sqrt(sum((row-mean(row))^2)/(length(row)-1))/sqrt(length(row))
}

se <- apply(rms,1,calc.se)
mean <- apply(rms,1,mean)


order.rms <- data.frame(order = factor(order, levels=order),
                        mean,
                        se,
                        rms)
order.rms
```

```r
melted <- melt(order.rms, variable='person', value.name='order.rms', id =c('order','mean','se'))

pdf(paste(wdir,'orderrms.pdf',sep=''), width=7, height=5)

p <- ggplot(data=melted, aes(x=factor(order), y=order.rms)) + theme_bw() 
p <- p + stat_summary(fun.y="mean", geom="bar")
p <- p + geom_errorbar(data = melted, aes(ymin = mean - se, ymax = mean + se, y = mean), 
                                       position = position_dodge(width = .9))

p
dev.off()
p
```
